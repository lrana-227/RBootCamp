```python
# Dependencies
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
```

```python
# Load crime data set into pandas
crime_data = pd.read_csv("../Resources/crime_data.csv")
crime_data.head()
```

```python
# Generate a scatter plot of year versus violent crime rate
year = crime_data.iloc[:,0]
violent_crime_rate = crime_data.iloc[:,3]
plt.scatter(year,violent_crime_rate)

plt.xlabel('Year')
plt.ylabel('Violent Crime Rate')
plt.show()
```

```python
# Perform a linear regression on year versus violent crime rate
s,i,r,p,s_err = stats.linregress(year,violent_crime_rate)
```

```python
# Create equation of line to calculate predicted violent crime rate
fit = s*year+i
```

```python
# Plot the linear model on top of scatter plot 
year = crime_data.iloc[:,0]
violent_crime_rate = crime_data.iloc[:,3]
plt.scatter(year,violent_crime_rate)
plt.plot(year,fit,"--")
plt.xlabel('Year')
plt.ylabel('Violent Crime Rate')
plt.show()
```

```python
# Repeat plotting scatter and linear model for year versus murder rate

year = crime_data.iloc[:,0]
mur= crime_data.iloc[:,5]

s,i,r,p,s_err = stats.linregress(year,mur)
fit1 = s*year+i
plt.scatter(year,mur)
plt.plot(year,fit1,"--")
plt.xlabel('Year')
plt.ylabel('Mur Crime Rate')
plt.show()
```

```python
# Repeat plotting scatter and linear model for year versus aggravated assault
year = crime_data.iloc[:,0]
a= crime_data.iloc[:,10]

s,i,r,p,s_err = stats.linregress(year,a)
fit2 = s*year+i
plt.scatter(year,a)
plt.plot(year,fit2,"--")
plt.xlabel('Year')
plt.ylabel('Assult Crime Rate')
plt.show()
```

```python
# Generate a facet plot of all 3 figures
fig,(ax1,ax2,ax3)= plt.subplots(3,sharex=True)
ax1.scatter(year,violent_crime_rate)
ax1.plot(year,fit,"--")
ax1.set_ylabel("violent crime")
ax2.scatter(year,mur)
ax2.plot(year,fit1,"--")
ax2.set_ylabel("murder crime")
ax3.scatter(year,a)
ax3.plot(year,fit2,"--")
ax3.set_ylabel("assult crime")
```
