```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set x axis to numerical value for month
x_axis_data = np.arange(1,13,1)
x_axis_data
```

```python
# Average weather temp
points = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Plot the line
plt.plot(x_axis_data,points)
plt.show()
```

```python
# Convert to Celsius C = (F-32) * 0.56      CAN ALSO DO point_C = [(F-32)*0.56 for F in points]
points_C = []
for nums in points:
    c = (nums -32) * 0.56 
    points_C.append(c)
points_C
```

```python
# Plot using Celsius
plt.plot(x_axis_data,points_C )
plt.show()
```

```python
# Plot both on the same chart
plt.plot(x_axis_data,points)
plt.plot(x_axis_data,points_C )
plt.xlabel("Something")
plt.ylabel("Anything")
plt.show()
```
