```python
# Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create the X and Y axis lists
time = np.arange(0,130,10)
danger_drop_speeds = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
railgun_speeds = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
# Plot the charts and apply some styling
danger_drop, = plt.plot(time,danger_drop_speeds,marker = 's',color='red', label = "Danger Drop Speed")
railgun, = plt.plot(time,railgun_speeds,marker = 'o',color='green', label = "Rail Gun Speed")
```

```python
# Add labels to X and Y axes :: Add title
plt.xlabel("Time")
plt.ylabel("Speeds")
plt.title("Rail Gun and Danger Drop Speeds and Time ")
```

```python
# Set the limits for the X and Y axes
plt.xlim(0,140)
plt.ylim(0,100)
```

```python
# Create a legend for the chart
plt.legend(handles =(danger_drop,railgun), loc="best")
```

```python
# Add in a grid for the chart
plt.grid()
```
