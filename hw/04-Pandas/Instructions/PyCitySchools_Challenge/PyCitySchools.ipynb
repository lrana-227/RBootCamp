```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools=school_data['school_name'].count()
total_students=student_data['student_name'].count()
total_budget=school_data['budget'].sum()
average_math_score=round(student_data['math_score'].mean(),2)
average_reading_score=round(student_data['reading_score'].mean(),2)

#Average for Math Scores --> needs a loop
count_math_score = 0
for x in student_data['math_score']:
    if x >= 70:
        count_math_score = count_math_score +1  
passing_percent_math= round((count_math_score/total_students)*100,2)

#Average for Reading Scores --> needs a loop
count_reading_score = 0
for y in student_data['reading_score']:
    if y >= 70:
        count_reading_score = count_reading_score +1  
passing_percent_reading=round( (count_reading_score/total_students)*100,2)

#Average for Overall Scores --> needs a loop
count_overall_score = 0
for i,score in student_data.iterrows():
    if (student_data['reading_score'][i] >= 70) & (student_data['math_score'][i] >= 70):
        count_overall_score = count_overall_score +1  
passing_percent_overall=round( (count_overall_score/total_students)*100,2)

district_summary_df = pd.DataFrame({"Total Schools":[total_schools],"Total Students":[total_students],"Total Budget":[total_budget],
                                   "Average Math Score":[average_math_score],"Average Reading Score":[average_reading_score],"% Passing Math":[passing_percent_math],
                                   "% Passing Reading":[passing_percent_reading],"% Overall Passing":[passing_percent_overall]})
district_summary_df

```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Groupby School to get the stats for all the values 
school_name = school_data_complete.groupby(["school_name"])

#Values to determine
school_type = school_name["type"].first()
total_students= school_name["student_name"].size()
total_school_budget= school_name["budget"].first()
per_student_budget= total_school_budget/total_students
average_math_score=school_name["math_score"].mean()
average_reading_score=school_name["reading_score"].mean()
percent_passing_math= ((student_data[student_data["math_score"]>=70.0].groupby(["school_name"]).size())/total_students)*100
percent_passing_reading= ((student_data[student_data["reading_score"]>=70.0].groupby(["school_name"]).size())/total_students)*100
percent_passing_overall= (student_data[(student_data["reading_score"]>=70.0) & (student_data["math_score"]>=70.0)].groupby("school_name")["Student ID"].count()/total_students)*100

school_summary_df = pd.DataFrame({
    
                    "School Type":school_type,
                    "Total Students":total_students,
                    "Total School Budget":total_school_budget,
                    "Per Student Budget":per_student_budget,
                    "Average Math Score":average_math_score,
                    "Average Reading Score":average_reading_score,
                    "% Passing Math":percent_passing_math,
                    "% Passing Reading":percent_passing_reading,
                    "% Overall Passing":percent_passing_overall})
school_summary_df


```
